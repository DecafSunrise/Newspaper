```python
def all_in_one(text, doc_id):
    outlist = []

    paragraphs = [sent for sent in re.split("[\n]{2,}", text) if sent != '']
    for para in paragraphs:
        para_id = "para://"+str(uuid.uuid4())
        sentences = sent_tokenize(para)
        for sent in sentences:
            sent_id = "sent://"+str(uuid.uuid4())
            outlist.append([doc_id, para_id, sent_id, sent])
            
    return outlist
```

```python
df['decomp'] = df.apply(lambda x: all_in_one(x.text, x.doc_id), axis=1)
```

```python
df['decomp'] = df['text'].apply(all_in_one)
```
